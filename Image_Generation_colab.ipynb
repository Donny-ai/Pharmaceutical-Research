```python
# loading stable diffusion model
from diffusers import StableDiffusionXLPipeline
import torch
pipe = StableDiffusionXLPipeline.from_pretrained('segmind/SSD-1B', torch_dtype=torch.float16, use_safetensors=True, variant='fp16')
pipe.to('cuda')
```

```python
# Logo prompting
prompt = 'orange logo of "f" with circular background, happy, calm, elegant, simple, soft, white background'
neg_prompt = '3D, complicated, sharp, extravagant, medieval, fancy, hard edges, angry, sad, arabic, eastern'
image = pipe(prompt=prompt, negative_prompt = neg_prompt).images[0]
```

```python
image.save('new F.png')
```
